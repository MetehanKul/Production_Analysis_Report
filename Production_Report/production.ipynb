```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_excel("Üretim.xlsx")
df.head(20)
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe().T
```

```python
print(df.columns)
df.columns = df.columns.str.replace(' ', '_').str.replace('İ', 'I').str.upper()

def replace_specific_vowels(column_name):
    vowels = "iöüİÖÜ"
    replacements = "ıouIOU"
    return column_name.translate(str.maketrans(vowels, replacements))

# Tüm kolon isimlerine bu dönüşümü uygulayalım
df.columns = [replace_specific_vowels(col) for col in df.columns]

df.head()
```

```python
df.isnull().sum().sum()
```

## Data Visualization

```python
df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['URETILEN_MIKTAR'], bins=30, kde=True, color='dodgerblue')

plt.title('Üretilen Miktar Dağılımı', fontsize=16, weight='bold')
plt.xlabel('Üretilen Miktar', fontsize=12)
plt.ylabel('Frekans', fontsize=12)

plt.show()
```

```python
# Scatter plot ile üretim miktarı ve toplam kar ilişkisi
plt.figure(figsize=(8, 5))
sns.regplot(x='URETILEN_MIKTAR', y='TOPLAM_KAR', data=df, scatter_kws={'color':'purple', 'alpha':0.6}, line_kws={'color':'red'})


plt.title('Üretim Miktarı ve Toplam Kar İlişkisi', fontsize=16, weight='bold')
plt.xlabel('Üretilen Miktar', fontsize=12)
plt.ylabel('Toplam Kar', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Violin plot ile Vardiya bazında üretim miktarı analizi
plt.figure(figsize=(8, 5))
sns.violinplot(x='VARDIYA', y='URETILEN_MIKTAR', data=df, palette='coolwarm')

plt.title('Vardiya Bazında Üretilen Miktar Yoğunluğu', fontsize=16, weight='bold')
plt.xlabel('Vardiya', fontsize=12)
plt.ylabel('Üretilen Miktar', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='VARDIYA', y='URETILEN_MIKTAR', data=df, estimator=sum, ci=None, palette='viridis')

for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.1f'), 
                     (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha = 'center', va = 'center', 
                     xytext = (0, 9), textcoords = 'offset points')


plt.title('Ekip Bazında Toplam Üretim Miktarı', fontsize=16, weight='bold')
plt.xlabel('Ekip', fontsize=12)
plt.ylabel('Toplam Üretilen Miktar', fontsize=12)

plt.show()
```

```python
# Sayısal değişkenler arasındaki korelasyonu hesaplayalım
plt.figure(figsize=(10, 6))
corr = df[['URETILEN_MIKTAR', 'ISKARTA_MIKTARI', 'TESLIM_ADEDI', 'ADET_FIYAT', 'KAR', 'ZARAR', 'TOPLAM_KAR']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=.5)

plt.title('Sayısal Değişkenler Arasındaki Korelasyon', fontsize=16, weight='bold')
plt.tight_layout()
plt.show()
```

```python
# Sayısal değişkenlerin pairplot analizi
sns.pairplot(df[['URETILEN_MIKTAR', 'ISKARTA_MIKTARI', 'TESLIM_ADEDI', 'TOPLAM_KAR']], kind='reg', diag_kind='kde', palette='red')

plt.suptitle('Sayısal Değişkenler Arası İlişki ve Dağılımlar', y=1.02, fontsize=16, weight='bold')
plt.tight_layout()
plt.show()
```

```python
# FacetGrid ile Vardiya bazında Üretilen Miktar
g = sns.FacetGrid(df, col="VARDIYA", height=4, aspect=1.5)
g.map(sns.histplot, 'URETILEN_MIKTAR', bins=20, kde=True)

plt.subplots_adjust(top=0.85)
g.fig.suptitle('Vardiya Bazında Üretilen Miktar Dağılımı', fontsize=16, weight='bold')
plt.show()
```

```python
# Vardiya bazında toplam karı hesaplayalım
ciro_per_ekip = df.groupby('VARDIYA')['TOPLAM_KAR'].sum().reset_index()

# Vardiya bazında en çok ciro yapan ekibi bulalım
max_ciro_ekip = ciro_per_ekip.loc[ciro_per_ekip['TOPLAM_KAR'].idxmax()]

print("En çok ciro yapan ekip:", max_ciro_ekip['VARDIYA'])
print("Toplam ciro:", max_ciro_ekip['TOPLAM_KAR'])
```

```python
# Vardiya bazında toplam ciroyu hesaplayalım
ciro_per_ekip = df.groupby('VARDIYA')['TOPLAM_KAR'].sum().reset_index()

plt.figure(figsize=(8, 5))
barplot = sns.barplot(x='VARDIYA', y='TOPLAM_KAR', data=ciro_per_ekip, palette='Blues')

for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.2f'), 
                     (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha='center', va='center', xytext=(0, 8), textcoords='offset points')

plt.title('Vardiya Bazında Toplam Ciro (Kar)', fontsize=16, weight='bold')
plt.xlabel('Vardiya', fontsize=12)
plt.ylabel('Toplam Ciro (Kar)', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Her ekip ve ürün kodu bazında toplam üretim miktarını hesaplayalım
urun_ekip_gruplama = df.groupby(['VARDIYA', 'URUN_KODU'])['URETILEN_MIKTAR'].sum().reset_index()

# Her ekip için en fazla üretilen ürün kodunu bulalım
max_uretim_urun = urun_ekip_gruplama.loc[urun_ekip_gruplama.groupby('VARDIYA')['URETILEN_MIKTAR'].idxmax()]

print(max_uretim_urun)
```

```python
# En fazla üretilen ürün kodlarını görselleştirelim
plt.figure(figsize=(8, 5))
sns.barplot(x='VARDIYA', y='URETILEN_MIKTAR', hue='URUN_KODU', data=max_uretim_urun, palette='Set1')

for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height():.2f}', 
                       (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha='center', va='center', xytext=(0, 8), textcoords='offset points')

plt.title('Her Ekip için En Fazla Üretilen Ürün Kodu', fontsize=16, weight='bold')
plt.xlabel('Ekip', fontsize=12)
plt.ylabel('Üretim Miktarı', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Her operatör bazında toplam üretim miktarını hesaplayalım
operatör_uretim = df.groupby('GOREVLI_OPERATOR')['URETILEN_MIKTAR'].sum().reset_index()

# Üretim miktarına göre en fazla üretim yapan ilk 15 operatörü sıralayalım
en_fazla_ureten_15_operatör = operatör_uretim.sort_values(by='URETILEN_MIKTAR', ascending=False).head(15)


print(en_fazla_ureten_15_operatör)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='URETILEN_MIKTAR', y='GOREVLI_OPERATOR', data=en_fazla_ureten_15_operatör, palette='coolwarm')

plt.title('En Fazla Üretim Yapan 15 Operatör', fontsize=16, weight='bold')
plt.xlabel('Toplam Üretilen Miktar', fontsize=12)
plt.ylabel('Operatör', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
en_fazla_ıskarta_operatör = df.groupby('GOREVLI_OPERATOR')['ISKARTA_MIKTARI'].sum().reset_index().sort_values(by = "ISKARTA_MIKTARI" , ascending=False).head(15)

plt.figure(figsize=(10, 6))
sns.barplot(x='ISKARTA_MIKTARI', y='GOREVLI_OPERATOR', data=en_fazla_ıskarta_operatör, palette='coolwarm')

plt.title('En Fazla Iskarta Yapan 15 Operatör', fontsize=16, weight='bold')
plt.xlabel('Toplam Iskarta Miktar', fontsize=12)
plt.ylabel('Operatör', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
vardiya_iskarta = df.groupby('VARDIYA')['ISKARTA_MIKTARI'].sum().reset_index()


en_fazla_iskarta_vardiya = vardiya_iskarta.loc[vardiya_iskarta['ISKARTA_MIKTARI'].idxmax()]

print("En fazla ıskarta veren vardiya:", en_fazla_iskarta_vardiya['VARDIYA'])
print("Toplam ıskarta miktarı:", en_fazla_iskarta_vardiya['ISKARTA_MIKTARI'])
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='VARDIYA', y='ISKARTA_MIKTARI', data=vardiya_iskarta, palette='Reds')

for i in plt.gca().patches:
    plt.gca().annotate(f'{i.get_height():.2f}', 
                       (i.get_x() + i.get_width() / 2., i.get_height()), 
                       ha='center', va='center', xytext=(0, 8), textcoords='offset points')


plt.title('Vardiyalara Göre Toplam Iskarta Miktarı', fontsize=16, weight='bold')
plt.xlabel('Vardiya', fontsize=12)
plt.ylabel('Toplam Iskarta Miktarı', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
urun_uretim = df.groupby('URUN_ADI')['URETILEN_MIKTAR'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='URETILEN_MIKTAR', y='URUN_ADI', data=urun_uretim, palette='coolwarm')

plt.title('Ürün Bazında Toplam Üretim Miktarı', fontsize=16, weight='bold')
plt.xlabel('Toplam Üretilen Miktar', fontsize=12)
plt.ylabel('Ürün Adı', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
urun_iskarta = df.groupby('URUN_ADI')['ISKARTA_MIKTARI'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='ISKARTA_MIKTARI', y='URUN_ADI', data=urun_iskarta, palette='Reds')

plt.title('Ürün Bazında Toplam Iskarta Miktarı', fontsize=16, weight='bold')
plt.xlabel('Toplam Iskarta Miktarı', fontsize=12)
plt.ylabel('Ürün Adı', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
urun_fiyatı = df.groupby('URUN_ADI')['ADET_FIYAT'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='ADET_FIYAT', y='URUN_ADI', data=urun_fiyatı, palette='Blues')

plt.title('Ürünlerin Adet Fiyatı', fontsize=16, weight='bold')
plt.xlabel('Adet Fiyatı', fontsize=12)
plt.ylabel('Ürün Adı', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
urun_adına_miktar= df.groupby(['URUN_ADI' , "VARDIYA"])['URETILEN_MIKTAR'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='URETILEN_MIKTAR', y='URUN_ADI', hue = "VARDIYA" , data=urun_adına_miktar, palette='Greens')

plt.title('Vardiyaların Toplam Ürettikleri Ürün Miktarı' , fontsize=16, weight='bold')
plt.xlabel('Ürün Miktarı', fontsize=12)
plt.ylabel('Ürün Adı', fontsize=12)

plt.tight_layout()
plt.show()
```
